### Bar Plot

```
plt.bar(df.region, df.crime_rate)
plt.xlabel("region")
plt.ylabel("crime rate")
plt.title("Crime Rate in Every Region")
plt.show()
```

```
company = ['Tokopedia', 'Bukalapak', 'Shopee', 'Akulaku', 'OLX']
revenue = [55, 40, 60, 47, 50]
profit =  [80, 50, 95, 65, 70]
list_com = np.arange(len(company))
list_com
```

```
plt.figure(figsize=(10,5))
plt.xticks(list_com, company) #change the xticks by it index and name
plt.bar(list_com-0.2, revenue, color='g', label='revenue', width=0.4)
plt.bar(list_com+0.2, profit, color='orange', label='profit', width=0.4)
plt.ylabel("Revenue")
plt.title("US Tech Stocks")
plt.legend()
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.yticks(list_com, company) #change the xticks by it index and name
plt.barh(list_com-0.2, revenue, color='g', label='revenue', height=0.4)
plt.barh(list_com+0.2, profit, color='orange', label='profit', height=0.4)
plt.ylabel("Revenue")
plt.title("US Tech Stocks")
plt.legend()
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.xticks(list_com, company) #change the xticks by it index and name
plt.bar(list_com, revenue, color='g', label='revenue', width=0.4)
plt.bar(list_com, profit, color='orange', label='profit', width=0.4, bottom=revenue)
plt.ylabel("Revenue")
plt.title("US Tech Stocks")
plt.legend()
plt.show()
```

### Pie Chart

```
firms = {"Frima A", "Firam B", "Firma C", "Firma D", "Firma E"}
markeet_share = [10, 5, 7, 14, 3] 
plt.figure(figsize=(8,5))
plt.pie(markeet_share, explode=[0, 0, 0, 0.1, 0], labels=firms, 
        autopct='%0.2f%%', shadow=False, startangle=-60, radius=3)
plt.legend(title='List of Firms')
plt.axis('equal')
plt.show()
```

### Multiple Plot
- nrows : number of rows in the figure
- ncols : number of columns in the figure
- index : index the plot

1 rows 2 columns

```
#first method

plt.figure(figsize=(10,5))
plt.subplot(1,2,1) #row, col, idx
plt.plot(df.work_force, df.income, 'go')
plt.title('Income VS Work Force')

plt.subplot(1,2,2)
plt.plot(df.hospital_beds, df.income, 'r*')
plt.title("Income VS Hospital Beds")

plt.suptitle("Sub Plot")
plt.show()
```

```
#second method

fig, (ax1,ax2) = plt.subplots(1,2, ) #1 row 2 col
fig.set_figheight(5)
fig.set_figwidth(10)
fig.suptitle("Sub Plot")

ax1.plot(df.work_force, df.income, 'go')
ax1.set_title('Income VS Work Force')

ax2.plot(df.hospital_beds, df.income, 'r*')
ax2.set_title("Income VS Hospital Beds")

plt.show()
```

2 rows 1 columns

```
plt.figure(figsize=(7,8))
plt.subplot(2,1,1) #rows, col, index
plt.plot(df.work_force, df.income, 'go')
plt.title('Income VS Work Force')

plt.subplot(2,1,2)
plt.plot(df.hospital_beds, df.income, 'r*')
plt.title("Income VS Hospital Beds")

plt.suptitle("Sub Plot")
plt.show()
```

2 rows 2 columns

```
fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,0].plot(df.hospital_beds, df.income, 'go')
ax[0,0].set_title("Subplot 1,1")

ax[0,1].plot(df.work_force, df.income, 'r^')
ax[0,1].set_title("Subplot 1,2")

ax[1,0].plot(df.percent_senior, df.income, 'b*')
ax[1,0].set_title("Subplot 2,1")

ax[1,1].plot(df.physicians, df.income, 'yx')
ax[1,1].set_title("Subplot 2,2")

plt.suptitle("Sub Plot 2 rows 2columns")
plt.show()
```

```
df.describe()
```


- Here as you can notice the mean value is less than the median value of each column. Median is represented by 50%(50th percentile) in the index column.This signifies the presence of Outliers. For example, a data set includes values: 30, 31, 32, and 2. The mean value (23.75), which is lower than the median of the data (30.5), is greatly affected by the extreme data point(2). 
- There is notably a large difference between 75th %tile and max values of predictors “residual sugar”, ” free sulfur dioxide”, ” total sulfur dioxide”. This indicates that some values of these 3 variables lie much farther from the general range of values( up to 75th %tile) 
- Thus, the observations 1 and 2 suggest that there are extreme values i.e Outliers in our dataset.

```
chicago_data['real_date'] = pd.to_datetime(chicago_data['Date'])
chicago_data.drop('Date', axis=1, inplace=True)
chicago_data.head()
```

```
def get_moth(x):
  return x.month

import calendar 
def get_day_name(x):
  return calendar.day_name[x.weekday()]
```

```
chicago_data["month"] = chicago_data["real_date"].apply(get_moth)
chicago_data["day"] = chicago_data["real_date"].apply(get_day_name)
chicago_data.head()
```

Visualization using Seaborn

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Scatter Plot

```
sns.scatterplot(df.alcohol, df.density)
plt.show()
```

Histogram

```
df.hist(figsize=(15,13))
plt.show()
```

```
sns.histplot(df.fixed_acidity, bins=10, kde=True)
plt.show()
```

Correlation

```
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap= "RdYlBu", annot=True)
plt.show()
```

- Alcohol has the highest positive correlation with wine quality, followed by the various other variables such as acidity, sulphates, density & chlorides
- There is a relatively high positive correlation between fixed_acidity and citric_acid, fixed_acidity and density.
-There is a relatively high negative correlation between fixed_acidity and pH.
-Density has a strong positive correlation with fixed_acidity, whereas it has a strong negative correlation with alcohol.
- Citric acid & volatile acidity have negative correlation.
- Free sulphur dioxide & total sulphur dioxide have positive correlation.


Count Plot

```
plt.figure(figsize=(10,5))
sns.countplot(df.quality)
plt.show()
```

Pair Plot

```
sns.pairplot(df)
plt.show()
```

- The correlation between fixed_acidity and citric_acid is 0.67 (you could find this value under the correlation matrix topic).
-Looking at scatterplot for this pair of variables, we can see the positive linear correlation between these two variables. We can observe the upward trend, and also the points are not too dispersed.

Box Plot

```
sns.boxplot(df.quality, df.alcohol, palette='GnBu_d')
plt.title("Box Plot Alcohol vs Quality")
plt.show()
```
